```python

def Get_PDF_File_As_String(filepath):
    fullText = []
    try: 
        with open(filepath, "rb") as f:
            pdf = pdftotext.PDF(f)
            for page in pdf:
                fullText.append(page)
        return '\n'.join(fullText)
    except:
        return "PDF-Error"
    
def Get_WORD_DOCX_File_As_String(filepath):
    fullText = []
    try:
        doc = docx.Document(filepath)
        fullText = []
        for para in doc.paragraphs:
            fullText.append(para.text)
        rtn_text = '\n'.join(fullText)
        return rtn_text
    except:
        return "DOCX-Error"
```

```python
filedata = []
filelist = []

for root, dirs, files in os.walk(root_folder):
    for file in files:
        #append the file name to the list
        filelist.append(os.path.join(root,file))


#print all the file names
for filepath in filelist:
    fp_split = filepath.split('/')
    file_name = fp_split[len(fp_split)-1].lower()
    file_folder = fp_split[len(fp_split)-2]
    #print("FName=,",file_name)
    if ".pdf" in file_name:
        #print("PDF_Content----------------",filepath)
        pdf_txt = Get_PDF_File_As_String(filepath)
        pdf_txt = pdf_txt.replace("\n", "")
        #pdf_txt = re.sub(r'[^\x00-\x7F]', '', pdf_txt)
        #pdf_txt = pdf_txt.encode("ascii","ignore")
        #pdf_txt = pdf_txt.decode()
        pdf_txt = pdf_txt[0:10000]
        #print("PDF Len=",str(len(pdf_txt)))
        #print("END_PDF_Content----------------")
        filedata.append([file_folder, pdf_txt])
    elif ".docx" in file_name:
        xyz = 0
        docx_txt = Get_WORD_DOCX_File_As_String(filepath)
        docx_txt = docx_txt.replace("\n", "")
        #docx_txt = re.sub(r'[^\x00-\x7F]', '', docx_txt)
        #docx_txt = docx_txt.encode("ascii","ignore")
        #docx_txt = docx_txt.decode()
        docx_txt = docx_txt[0:10000]
        filedata.append([file_folder,docx_txt])

file_df = pd.DataFrame(filedata)
file_df.columns = ['doc_type','doc_text']
file_df['doc_text'] = file_df['doc_text'].astype('str')
file_df   
```

```python
# remove rows with text len of less than 20 characters
file_df = file_df[file_df['doc_text'].str.len() > 20]
file_df.reset_index()
file_df 
```
